```python
import pickle
import pandas as pd
import numpy as np
```

```python


# Listado de los archivos pkl para cada estrategia
files1 = [
    # Archivos para df_strategy_1
    'df_strategy_1_100_125_150.pkl',
    'df_strategy_1_100_150_125.pkl',
    'df_strategy_1_100_75_50.pkl',
    'df_strategy_1_100_50_75.pkl',
    'df_strategy_1_100_100_100.pkl',

    # Archivos para df_strategy_2
    'df_strategy_2_100_125_150.pkl',
    'df_strategy_2_100_150_125.pkl',
    'df_strategy_2_100_75_50.pkl',
    'df_strategy_2_100_50_75.pkl',
    'df_strategy_2_100_100_100.pkl',

    # Archivos para df_strategy_3
    'df_strategy_3_100_125_150.pkl',
    'df_strategy_3_100_150_125.pkl',
    'df_strategy_3_100_75_50.pkl',
    'df_strategy_3_100_50_75.pkl',
    'df_strategy_3_100_100_100.pkl',

    # Archivos para df_strategy_4
    'df_strategy_4_100_125_150.pkl',
    'df_strategy_4_100_150_125.pkl',
    'df_strategy_4_100_75_50.pkl',
    'df_strategy_4_100_50_75.pkl',
    'df_strategy_4_100_100_100.pkl'
]

files = ['data/'+f for f in files1]

# Lista para almacenar los resultados
results = []

# Iterar sobre cada archivo pickle
for file in files:
    # Cargar el DataFrame desde el archivo pickle
    with open(file, 'rb') as f:
        df = pickle.load(f)
    
    # Agrupar por la columna 'prob_symp_more_5d' y calcular la media de 'total_cost'
    grouped_df = np.round(df.groupby('prob_symp_more_5d')['total_cost'].mean().reset_index(),2)
    
    # Crear un identificador único para cada conjunto de C1, C2, C3
    # Asumiendo que el nombre del archivo refleja los valores de C1, C2, C3
    params = file.replace('data/df_strategy_', '').replace('.pkl', '').split('_')
    strategy = params[0]  # Extraemos la estrategia (df_strategy_1, df_strategy_2, etc.)
    C1, C2, C3 = map(int, params[1:4])  # Extraemos C1, C2, C3

    # Añadir las columnas strategy, C1, C2, C3 a los resultados
    for _, row in grouped_df.iterrows():
        results.append({
            'strategy': strategy,
            'C1': C1,
            'C2': C2,
            'C3': C3,
            'prob_symp_more_5d': row['prob_symp_more_5d'],
            'total_cost': row['total_cost']
        })

# Convertir los resultados a un DataFrame
result_df = pd.DataFrame(results)

# Mostrar el DataFrame resultante
result_df=result_df.sort_values(['C1','C2','C3','prob_symp_more_5d','strategy'])
#result_df = result_df[result_df['C2']!=100]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
result_df['C_comb'] =np.concatenate([np.repeat(i, 12) for i in range(1, 6)])
result_df["total_cost_k"] = result_df["total_cost"] / 1000

g = sns.FacetGrid(result_df, col="prob_symp_more_5d", row="C_comb", height=3, aspect=1, palette="tab10",)

# Mapear el gráfico de barras
g.map(sns.barplot, "strategy","total_cost_k", fill=True, errorbar=None, palette="tab10")

# Mejorar la visualización con títulos, etiquetas y leyenda
g.set_axis_labels("Strategy", "Total cost (in thousands)", fontsize=12)
g.set_titles("{col_name}, Cost combination: {row_name}", fontsize=32)
g.add_legend()

for ax in g.axes.flat:
    ax.title.set_fontsize(12)  # Ajusta el tamaño del título de cada faceta

# Ajustar el diseño para evitar superposiciones
plt.tight_layout()
plt.savefig("img_sensitivity/bar_total_cost_strategies.pdf")
plt.show()
```

```python

# Crear el gráfico
g = sns.relplot(
    data=result_df,
    x="strategy", y="total_cost_k",
    hue="C_comb", col="prob_symp_more_5d",
    kind="line",
    height=5, aspect=.75, facet_kws=dict(sharex=False), marker="o",palette='Set2'
)

# Establecer las etiquetas de los ejes
g.set_axis_labels("Strategy", "Total cost (in thousands)", fontsize=18)

# Establecer los títulos de cada facet (ajustado correctamente)
g.set_titles("{col_name:.0%}", pad=2)
g.axes[0, 0].set_xlabel("")
g.axes[0, 1].set_xlabel("Strategy")
g.axes[0, 2].set_xlabel("")

# Ajustar el tamaño de la fuente de los títulos de las facetas
for ax in g.axes.flat:
    ax.title.set_fontsize(18)

# Ajustar el tamaño de las etiquetas de los ejes x y y
for ax in g.axes.flat:
    ax.tick_params(axis='x', labelsize=18)  # Cambiar el tamaño de las etiquetas del eje x
    ax.tick_params(axis='y', labelsize=18)  # Cambiar el tamaño de las etiquetas del eje y

# Agregar leyenda y ajustar el tamaño de la fuente

legend = g._legend
legend.set_title("Cost Combination")  # Cambiar el tamaño del título
for text in legend.get_texts():
    text.set_fontsize(16)  # Cambiar el tamaño del texto de la leyenda

legend.get_title().set_fontsize(16)  #

# Ajustar la posición de la leyenda para evitar que se sobreponga
g.legend.set_bbox_to_anchor((1.15, 0.5))  # Mueve la leyenda fuera del gráfico

# Ajustar el diseño para evitar superposiciones
plt.tight_layout()

# Guardar la imagen
plt.savefig("img_sensitivity/line_total_cost_strategies.pdf", bbox_inches='tight')

# Mostrar el gráfico
plt.show()
```
