```python
def plot(df, variable="total_cost"):
    feature_title = {"total_cost": "total cost"}

    if variable not in feature_title.keys():
        raise NameError(
            "Variable " + variable + " is not valid.\n"
            "The variable must be one of: " + ", ".join(list(feature_title.keys())),
        )

    def vertical_mean_line(x, **kwargs):
        plt.axvline(x.mean(), linestyle="--", color="blue", linewidth=3)

    strategy = df.strategy[0]
    people = 1000
    model_parameters = pd.read_pickle(filepath_or_buffer="data/model_parameters.pkl")
    if strategy == 4:
        g = sns.displot(df, x=variable)
        g.map(vertical_mean_line, variable)
        # g.refline(x=df[variable].mean(), color="blue", lw=3)
        g.refline(x=people * 100, color="red", lw=3)
        plt.show()
    else:
        g = sns.displot(df, x=variable, col="prob_symp_more_5d")
        # g.map(vertical_mean_line, variable)
        # g.refline(x=df[variable].mean(), color="blue", lw=3)
        # g.refline(x=people * 100, color="red", lw=3)
        plt.show()
```

```python
import numpy as np

df_all_melt = df_all.melt(
    id_vars=["strategy", "prob_symp_more_5d"],
    value_vars=["number_positive_reported", "number_test_per_person", "total_cost_k"],
    var_name="variable",
    value_name="value",
)

h = 10
w = h * 1.618
# Create plot using seaborn
sns.set(style="ticks")
sns.set_context("paper", font_scale=1.7)
plt.figure(figsize=(w, h))  # set the figure size

g = sns.displot(
    df_all_melt,
    x="value",
    col="prob_symp_more_5d",
    row="variable",
    hue="strategy",
    kind="kde",
    fill=True,
    # binwidth=0.01,
    # stat="probability",
    palette="tab10",
    height=3,
    aspect=1,
    facet_kws={"sharey": False, "sharex": False},
)


set_titles = g.set_titles("{col_name:.0%}", pad=2)

axes = g.axes.flatten()

axes[1].set_title("Distribution of positive reported\n\n50%")
axes[4].set_title("\nDistribution of tests per person\n\n50%")
axes[7].set_title("\nDistribution of total cost (in thousands)\n\n50%")

for ax in axes:
    ax.set_xlabel("")
    ax.set_ylabel("")
    ax.set_yticks([])

for i in range(len(axes)):
    # Get current x-axis ticks for the third facet (flattened index 2)
    current_ticks = axes[i].get_xticks()
    # Create a new set of ticks with more granularity
    new_ticks = np.linspace(current_ticks[0], current_ticks[-1], num=5)
    axes[i].set_xticks(new_ticks)

for i in [6, 7, 8]:
    current_tick_labels = axes[i].get_xticklabels()
    new_tick_labels = [
        f"{int(float(label.get_text())):,.0f}" for label in current_tick_labels
    ]
    axes[i].set_xticklabels(new_tick_labels)

for i in [3, 4, 5]:
    current_tick_labels = axes[i].get_xticklabels()
    new_tick_labels = [
        f"{float(label.get_text()):,.1f}" for label in current_tick_labels
    ]
    axes[i].set_xticklabels(new_tick_labels)


plt.tight_layout()
sns.move_legend(g, "center right", bbox_to_anchor=(1.1, 0.5), title="Strategy")
plt.savefig("img/plot_uhist_montecarlo_strategies.pdf", bbox_inches="tight")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


df_all_strategy_sensitivity = pd.read_pickle(
    filepath_or_buffer="data/df_all_strategy_sensitivity.pkl"
)

# group by strategy and prob_symp_more_5d and summarize efficiency and stock_capacity with mean
df_strategy_sensitivity = (
    df_all_strategy_sensitivity.groupby(
        ["strategy", "prob_symp_more_5d", "cost_antigen", "cost_lamp"], as_index=False
    )
    # estimate the mean of efficiency and stock_capacity
    .agg(
        {
            "total_cost": "mean",
            "efficiency": "mean",
            "stock_capacity": "mean",
            "number_positive_reported": "mean",
            "number_test_per_person": "mean",
        }
    )
    # filter out rows where prob_symp_more_5d is 0.5
    .query("prob_symp_more_5d == 0.5")
    .query("strategy == 4")
    # sort by cost_antigen and cost_lamp
    # .sort_values(by=["cost_antigen", "cost_lamp"])
    .reset_index()
)

# df_strategy_sensitivity["total_cost_k"] = df_strategy_sensitivity["total_cost"] / 1000
# df_strategy_sensitivity["C_comb"] = (
#     df_strategy_sensitivity["cost_antigen"].astype(str)
#     + "_"
#     + df_strategy_sensitivity["cost_lamp"].astype(str)
# )


g = sns.lineplot(
    data=df_strategy_sensitivity,
    x="cost_antigen",
    y="efficiency",
    # hue="strategy",
    size="cost_lamp",
    # style="strategy",
    markers=True,
    palette="tab10",
    errorbar=None,
)

# set title legend
g.set_title("Sensitivity of Efficiency against changes in test costs", fontsize=14)
g.set_xlabel("Cost of Antigen Test", fontsize=12)
g.set_ylabel("Efficiency", fontsize=12)
# set legend caption
g.legend(title="Cost of RT-LAMP", title_fontsize=12, fontsize=10)

# position the legend right center

# sns.move_legend(g, "center right", bbox_to_anchor=(1.5, 1), title="")
plt.tight_layout()

plt.savefig("img/plot_sensitivity_efficiency_vs_costs.pdf", bbox_inches="tight")
plt.show()
```
