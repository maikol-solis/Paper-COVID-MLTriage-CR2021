```python
import warnings

import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import pandas as pd
import seaborn as sns

warnings.filterwarnings("ignore")

#!pip install seaborn --upgrade
```

```python
def plot(df, variable="total_cost"):
    feature_title = {"total_cost": "total cost"}

    if variable not in feature_title.keys():
        raise NameError(
            "Variable " + variable + " is not valid.\n"
            "The variable must be one of: " + ", ".join(list(feature_title.keys())),
        )

    def vertical_mean_line(x, **kwargs):
        plt.axvline(x.mean(), linestyle="--", color="blue", linewidth=3)

    strategy = df.strategy[0]
    people = 1000
    model_parameters = pd.read_pickle(filepath_or_buffer="data/model_parameters.pkl")
    if strategy == 4:
        g = sns.displot(df, x=variable)
        g.map(vertical_mean_line, variable)
        # g.refline(x=df[variable].mean(), color="blue", lw=3)
        g.refline(x=people * 100, color="red", lw=3)
        plt.show()
    else:
        g = sns.displot(df, x=variable, col="prob_symp_more_5d")
        # g.map(vertical_mean_line, variable)
        # g.refline(x=df[variable].mean(), color="blue", lw=3)
        # g.refline(x=people * 100, color="red", lw=3)
        plt.show()
```

```python
df_strategy_1 = pd.read_pickle(filepath_or_buffer="data/df_strategy_1.pkl")
df_strategy_2 = pd.read_pickle(filepath_or_buffer="data/df_strategy_2.pkl")
df_strategy_3 = pd.read_pickle(filepath_or_buffer="data/df_strategy_3.pkl")
df_strategy_4 = pd.read_pickle(filepath_or_buffer="data/df_strategy_4.pkl")
df_all = pd.concat([df_strategy_1, df_strategy_2, df_strategy_3, df_strategy_4])
```
